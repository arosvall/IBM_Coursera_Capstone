## Problem Statement

A retail company wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month.

## Questions and Code

```python
import numpy as np
import pandas as pd


from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

np.random.seed = 42
```

Load the given dataset.

```python
data = pd.read_csv("sales_data.csv")
data.head()
```

**1. Randomly split the given data into 2 subsets for training (80%) and test (20%). Use *random_state = 42*. [2 points]**

```python
train_set, test_set = train_test_split(data, test_size = 0.2, random_state=42)
```

**2. Save the 2 subsets to 2 CSV files: ``sales_train.csv`` and ``sales_test.csv``. [2 points]**

```python
train_set.to_csv('sales_train.csv', index=False)
test_set.to_csv('sales_test.csv', index=False)
```

**3. Reload the two newly created CSV files into 2 dataframes ``train`` and ``test``. [2 points]**

```python
train_set.head()
```

```python
train = pd.read_csv('sales_train.csv')
test = pd.read_csv('sales_test.csv')

train.head()
```

**4. Convert all category data in ``train`` and ``test`` into numerical data. [4 points]**

```python
train.dtypes
```

```python
train = pd.get_dummies(train, prefix=['Age', 'City_Category', 'Gender', 'Stay_In_Current_City_Years'])
train.head()
```

```python
test = pd.get_dummies(test, prefix=['Age', 'City_Category', 'Gender', 'Stay_In_Current_City_Years'])
test.head()
```

**5. Build a Linear Regression model using ``train`` and all the columns except ``User_ID`` and ``Purchase`` as the predictors. Report RMSE values on the training and test sets. [3 points]**

Document: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
train_x = train.copy()
train_y = train['Purchase']
train_x.drop(['User_ID', 'Purchase'], axis=1, inplace=True)
train_x.head()

test_x = test.copy()
test_y = test['Purchase']
test_x.drop(['User_ID', 'Purchase'], axis=1, inplace=True)
test_x.head()
```

```python
from sklearn.linear_model import LinearRegression
reg_model = LinearRegression().fit(train_x, train_y)
reg_model_train_pred = reg_model.predict(train_x)
reg_model_test_pred = reg_model.predict(test_x)
```

```python
mse = mean_squared_error(train_y, reg_model_train_pred)
print("Train RMSE: {error}".format(error=np.sqrt(mse)))

mse = mean_squared_error(test_y, reg_model_test_pred)
print("Test RMSE: {error}".format(error=np.sqrt(mse)))
```

**6. Build a Decision Tree Regressor with default parameters using train and all the columns except ``User_ID`` and ``Purchase`` as the predictors. Report RMSE values on the training and test sets. [2 points]**

Document: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html#sklearn.tree.DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor().fit(train_x, train_y)
dt_reg_train_pred = dt_reg.predict(train_x)
dt_reg_test_pred = dt_reg.predict(test_x)
```

```python
mse = mean_squared_error(train_y, dt_reg_train_pred)
print("Train RMSE: {error}".format(error=np.sqrt(mse)))

mse = mean_squared_error(test_y, dt_reg_test_pred)
print("Test RMSE: {error}".format(error=np.sqrt(mse)))
```

**7. Use RandomizedSearchCV to tune parameters ``max_depth``, ``min_samples_split``, ``min_samples_leaf`` and ``max_features`` of the Decision Tree Regressor in order to obtain a lower RMSE value on the test set compared to that in Question 6. [5 points]**

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as rand_int

param_dist = {"max_depth": [3, 8],
              "max_features": rand_int(1, 11),
              "min_samples_split": rand_int(2, 11),
              "min_samples_leaf": rand_int(2,10)}

n_iter_s = 50

random_search = RandomizedSearchCV(dt_reg, param_distributions=param_dist,
                                   n_iter=n_iter_s, cv=5)

rs_dt = random_search.fit(train_x, train_y)
rs_dt_train_pred = rs_dt.predict(train_x)
rs_dt_test_pred = rs_dt.predict(test_x)
```

```python
mse = mean_squared_error(train_y, rs_dt_train_pred)
print("Train RMSE: {error}".format(error=np.sqrt(mse)))

mse = mean_squared_error(test_y, rs_dt_test_pred)
print("Test RMSE: {error}".format(error=np.sqrt(mse)))
```
